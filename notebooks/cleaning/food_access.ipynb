```python
import pandas as pd
import numpy as np
import math
```

```python
# Read in raw data
df = pd.read_csv("../../data/raw/FoodAccessResearchAtlasData2019.csv")
df
```

```python
# Create a column for county level fips
df["fips"] = (df["CensusTract"] / 1000000).apply(np.floor).astype(int)
```

```python
# Limit to columns of interest
df = df[["fips", "LALOWI1_10", "Pop2010"]]
# Group by fips to aggregate all tracts in each county
df = df.groupby("fips").sum().reset_index()
df
```

```python
# Divide out the population living in low income, low access areas by the county poulation
# to find the low income, low access share of the population
df["low_income_low_access_share"] = df["LALOWI1_10"] / df["Pop2010"]
df
```

```python
# Remove intermediate columns
df = df[["fips", "low_income_low_access_share"]]
df
```

```python
# Write out the cleaned data
df.to_csv("../../data/processed/food_access.csv", index=False)
```
